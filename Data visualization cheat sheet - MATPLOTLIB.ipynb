```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# functional method
X = np.linspace(0,10,11)
y = X**2
plt.plot(X,y,'r-') # last parameter is color and line style
plt.xlabel('lab')
plt.ylabel('lab')
plt.title('tit')
```

```python
# multiple plots
plt.subplot(1,2,1) # rows, columns 1st plot
plt.plot(X,y,'r')

plt.subplot(1,2,2) # rows, columns, 2nd plot
plt.plot(y,X,'b')
```

```python
# object oriented method

fig = plt.figure() # create a figure object
axes = fig.add_axes([0.5,0.5,0.8,0.8]) #left, bottom, width, height in the list
axes.plot(X,y)
axes.set_xlabel('string')
axes.set_ylabel('string')
axes.set_title('string')
```

```python
# for overlapping graphs

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.6,0.2,0.2])
axes1.plot(X,y)
axes2.plot(y,X)
axes1.set_title('test')
axes2.set_title('test2')
```

```python
# for multiple graphs

# axes is an array that is iterable

fig,axes = plt.subplots(2,2) # nrows,ncolumns
plt.tight_layout() # make it look better and less cluttered
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(X,y)
axes[0].set_title('first plot')
axes[1].plot(y,X)
axes[1].set_title('second plot')
```

```python
# figure size
# dpi is really cool

fig,axes = plt.subplots(nrows=1, ncols=2, figsize = (6,3),dpi=100)
axes[0].plot(X,y)
axes[1].plot(y,X)
```

```python
# saving a figure

fig.savefig('mypng.png',dpi=200)
```

```python
# random normal coordinates per point
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
# random uniform colours and sizes per point
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.figure(figsize=(8, 8))
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

```python
# Error Bars

index = range(10)
sample_means = []
sample_vars = []

for i in index:
    sample = np.random.normal(loc=i, scale=i, size=100)
    sample_means.append(np.mean(sample))
    sample_vars.append(np.var(sample))
    
plt.errorbar(index, sample_means, yerr=sample_vars, fmt='.k');
```

```python
plt.errorbar(index, sample_means, yerr=sample_vars, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```
